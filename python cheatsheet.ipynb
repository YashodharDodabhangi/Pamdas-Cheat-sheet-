```python
#Accesing the Columns in pandas
dataset[['volatile acidity','citric acid']] 
```

```python
#Accesing the rows or slicing the rows
# by numbering  of the rows as dataframe.loc[[start_row:end_row ]]
dataset.loc[0:3]  # starting 3 rows
```

```python
#check shape of data
dataset.shape
```

```python
#removing the rows or columns using drop()
dataset.drop([0,3,6],axis=0) #removing the 0th row 3rd and 6th row  and axis=0 implies by row

```

```python
#Removing the columns
dataset.drop(['fixed acidity','volatile acidity'],axis=1) #removing fixed acidity and volatile acidity column and axis=1 implies by row
```
